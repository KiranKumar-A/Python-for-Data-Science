# Pandas Built-in Data Visualization

In this lecture we will learn about pandas built-in capabilities for data visualization! It's built-off of matplotlib, but it baked into pandas for easier usage!  

Let's take a look!

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1')
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
df3 = pd.read_csv('df3')
df3.head()
```

```python
df1['A'].hist(bins = 30)
```

```python
df1['A'].plot(kind = 'hist', bins =20)
```

```python
df1['A'].plot.hist(bins = 30)
```

```python
df2.plot.area(alpha = 0.4)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)
```

```python
df1['A'].plot.hist(bins = 50)
```

```python
df1.plot.line(x = df1.index, y = 'B', figsize = (12,3), lw = 1)
```

```python
df1.plot.scatter(x = 'A', y = 'B', c= 'C')
```

```python
df1.plot.scatter(x = 'A', y = 'B', c= 'C', cmap = 'coolwarm')
```

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['C']*100)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns = ['a', 'b'])
```

```python
df.plot.hexbin(x = 'a', y = 'b')
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 20, cmap = 'coolwarm' )
```

```python
df2['a'].plot.density()
```

```python
df2.plot.density()
```
